### Задание 4 (1 балл)

Преобразуйте колонку `date_added` в datetime. Создайте новые столбцы month и year из преобразованного столбца. Постройте линейный график: сколько в среднем фильмов было снято в каждый из месяцев.

Добавьте на график:

+ Заголовок жирным шрифтом
+ Подписи осей
+ Маркеры на точках перегиба
+ Настройте ось X: ось начинается с 1 заканчивается 12, шаг 1

````python
from scipy.signal import find_peaks


df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['month'] = df['date_added'].dt.month
df['year'] = df['date_added'].dt.year
monthly_avg = df.groupby('month').size() / df['year'].nunique()
plt.figure(figsize=(10, 6))
plt.plot(monthly_avg.index, monthly_avg.values, marker='o', linestyle='-', color='b')

peaks, _ = find_peaks(monthly_avg.values)
valleys, _ = find_peaks(-monthly_avg.values)
plt.scatter(monthly_avg.index[peaks], monthly_avg.values[peaks], color='r', zorder=5, label='Peaks')
plt.scatter(monthly_avg.index[valleys], monthly_avg.values[valleys], color='r', zorder=5, label='Valleys')

plt.title("Среднее количество фильмов, выходящих в прокат Каждый месяц", fontsize=16, fontweight='bold')
plt.xlabel("Месяц", fontsize=12)
plt.ylabel("Среднее количество фильмов", fontsize=12)
plt.xticks(range(1, 13))
plt.tight_layout()
plt.show()
````
